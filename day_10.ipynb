```python
num = 12345673

length = len(str(num))
half = length // 2

first_half = num // (10 ** half)
remaining = num % (10 ** half)

n = first_half
rev = 0
while n > 0:
    rev = rev * 10 + (n % 10)
    n //= 10
print(f"{rev}{remaining}")
```
